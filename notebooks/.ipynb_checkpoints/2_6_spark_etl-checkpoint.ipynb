## ETL Multi-nivel: Landing → Processed (Bronze → Silver)

En esta sección implementaremos el proceso de transformación desde el nivel `landing` (datos crudos) al nivel `processed` (datos limpios y enriquecidos).

### Arquitectura de Capas

- **Landing (Bronze)**: Datos crudos, tal como llegan del origen. Particionados por año/mes, pueden contener duplicados o datos incompletos.
- **Processed (Silver)**: Datos limpios, validados y enriquecidos. Libres de duplicados y nulls innecesarios, con tipos de dato correctos.
- **Curated (Gold)**: Datos finales optimizados para análisis, agregaciones y reportes.

### Operaciones en la Capa Silver

En esta capa realizamos:
1. **Limpieza**: Eliminar duplicados, validar tipos de datos
2. **Enriquecimiento**: Joins con tablas dimensionales (employees, customers)
3. **Cálculos derivados**: Agregar columnas calculadas (comisiones, márgenes)
4. **Particionamiento**: Reorganizar según estrategia de consulta
5. **Validación**: Checks de calidad de datos

### Tracking de Procesamiento

Para evitar reprocesar datos y mantener la continuidad, usaremos un archivo de control (`_processing_status.json`) que registra:
- Último período procesado (año-mes)
- Timestamp de procesamiento
- Número de registros procesados
- Estado (success/failed)

De esta forma, al ejecutar el pipeline de nuevo, solo procesaremos datos nuevos desde el último estado registrado.

```python
import json
import os
from datetime import datetime

from pyspark.sql import DataFrame, Row

# Configuración de rutas ETL
PROCESSED_PATH: str = f"{DATALAKE_PATH}/processed"
PROCESSING_STATUS_FILE: str = f"{PROCESSED_PATH}/_processing_status.json"

# Crear directorio processed si no existe
if SPARK_MASTER != "yarn":
    os.makedirs(PROCESSED_PATH, exist_ok=True)


def read_processing_status() -> dict:
    """Lee el estado de procesamiento desde el archivo de control usando Spark.

    Usar Spark para leer el JSON permite tratar rutas locales y HDFS de la misma forma.
    Devuelve un dict con las claves esperadas o un valor por defecto si no existe.
    """
    try:
        # Intentar leer con Spark (funciona tanto en HDFS como en filesystem local)
        df: DataFrame = spark.read.json(PROCESSING_STATUS_FILE)

        rows: list[Row] = df.collect()
        if not rows:
            return {"last_processed_month": None, "history": []}

        # Tomamos la primera fila como el objeto de estado
        row: dict = rows[0].asDict()

        # Normalizar posibles filas internas (Rows) dentro de 'history'
        if "history" in row and isinstance(row["history"], list):
            row["history"] = [r.asDict() if hasattr(r, "asDict") else r for r in row["history"]]

        return row
    except Exception as e:
        print(f"Aviso: No se pudo leer status anterior: {e}")
        return {"last_processed_month": None, "history": []}


def write_processing_status(status: dict) -> None:
    """Escribe el estado de procesamiento al archivo de control usando Spark.

    Convierte el `status` a tipos JSON-serializables y lo escribe con Spark
    para que la lectura posterior con `spark.read.json` sea consistente
    tanto en HDFS como en filesystem local.
    """
    try:
        # Normalizar objetos no serializables (e.g., datetime) a strings
        safe_status = json.loads(json.dumps(status, default=str))

        # Crear DataFrame a partir del dict
        df: DataFrame = spark.createDataFrame([safe_status])

        target_path: str = PROCESSING_STATUS_FILE

        # Escribir en modo overwrite para reemplazar el estado anterior
        df.coalesce(1).write.mode("overwrite").json(target_path)

        print(f"Status guardado: {target_path}")
    except Exception as e:
        print(f"Error al guardar status con Spark: {e}")


# Leer estado previo
current_status: dict = read_processing_status()
print("Estado de procesamiento actual:")
print(json.dumps(current_status, indent=2, default=str))
```

Cargamos los dataframes necesarios desde *Landing*.

```python
# EMPLOYEES se carga desde el datalake
employees_parquet_file: str = f"{DATALAKE_PATH}/landing/employees"
dfEmpleados: DataFrame = (
    spark.read.format("parquet").option("mergeSchema", "true").load(employees_parquet_file)
)
dfEmpleados.cache()

# CUSTOMERS se carga desde el datalake
customers_parquet_file: str = f"{DATALAKE_PATH}/landing/customers"
dfClientes: DataFrame = (
    spark.read.format("parquet").option("mergeSchema", "true").load(customers_parquet_file)
)
dfClientes.cache()
```

Función para calcular las comisiones:

```python
from pyspark.sql import functions as F
from pyspark.sql.types import DoubleType


@F.udf(returnType=DoubleType())
def _recommended_commission(sale_amount: float, sale_date: datetime.date, hire_date: datetime.date):
    """
    Calcula la tasa de comisión recomendada basada en el importe de la venta
    y la antigüedad del vendedor:
    1. Comisión base del 3%
    2. Si la venta es mayor a 1000€, añadir 2%
    3. Si el vendedor lleva más de 6 meses en la empresa, añadir 1%
    """
    base: float = 0.03
    try:
        if sale_amount is None:
            sale_amount = 0.0
        # Determinamos si el empleado lleva contratado más de 6 meses a la fecha de la venta
        employed_longer: bool = False
        if hire_date is not None and sale_date is not None:
            # hire_date / sale_date pueden ser date o datetime
            sd: datetime.date = (
                sale_date if isinstance(sale_date, datetime.date) else sale_date.date()
            )
            hd: datetime.date = (
                hire_date if isinstance(hire_date, datetime.date) else hire_date.date()
            )
            employed_longer: bool = (sd - hd).days > 180

        if sale_amount > 1000:
            base += 0.02
        if employed_longer:
            base += 0.01
        return base
    except Exception:
        return base
```

### Paso 1: Procesar Ventas (Sales)

Leeremos desde `landing/sales_by_year_month`, aplicaremos transformaciones y escribiremos en `processed/sales_processed`.

**Transformaciones a aplicar:**
- Eliminar duplicados por SalesID
- Validar tipos de dato
- Derivar columna de ComissionRate (usando UDF anterior)
- Enriquecer con información de empleados y clientes
- Particionamiento por año y mes

```python
from pyspark.sql.functions import col, month
from pyspark.sql.functions import round as sql_round
from pyspark.sql.functions import year

# 2. Limpieza: Eliminar duplicados
sales_clean: DataFrame = sales.dropDuplicates(["SalesID"])
print(f"Registros tras eliminar duplicados: {sales_clean.count()}")

# 3. Enriquecimiento: Agregar información de empleados
sales_enriched: DataFrame = sales_clean.join(
    dfEmpleados.select("EmployeeID", "HireDate", "FirstName", "LastName"),
    sales_clean["SalesPersonID"] == dfEmpleados["EmployeeID"],
    how="left",
).drop(
    "EmployeeID"
)  # Evitar columna duplicada

# 4. Agregar comisión usando la UDF
sales_enriched = sales_enriched.withColumn(
    "CommissionRate", _recommended_commission(col("TotalPrice"), col("SalesDate"), col("HireDate"))
).withColumn("Commission", sql_round(col("TotalPrice") * col("CommissionRate"), 2))

# 5. Agregar columnas de particionamiento (año, mes)
sales_enriched = sales_enriched.withColumn("year", year(col("SalesDate"))).withColumn(
    "month", month(col("SalesDate"))
)

# 6. Seleccionar columnas finales en orden lógico
sales_processed: DataFrame = sales_enriched.select(
    col("SalesID"),
    col("SalesDate"),
    col("year"),
    col("month"),
    col("SalesPersonID"),
    col("FirstName").alias("EmployeeFirstName"),
    col("LastName").alias("EmployeeLastName"),
    col("HireDate"),
    col("CustomerID"),
    col("ProductID"),
    col("Quantity"),
    col("TotalPrice"),
    col("Discount"),
    col("CommissionRate"),
    col("Commission"),
)

print("\nSchema del DataFrame procesado:")
sales_processed.printSchema()
print(f"\nRegistros procesados: {sales_processed.count()}")
```

```python
sales_processed.show(5, truncate=False)
```

### Paso 2: Escribir en Processed con Particionamiento

Escribimos los datos particionados por **year** y **month**. Esto permite:
- Lecturas rápidas por rango de fechas
- Actualizaciones incrementales (agregar solo nuevos meses)
- Facilita la limpieza/reemplazo de particiones específicas

```python
from datetime import datetime

# Ruta de salida
sales_processed_path: str = f"{PROCESSED_PATH}/sales_processed"

# Escribir con particionamiento por year y month
# Usamos mode="overwrite" para reemplazar si existen las mismas particiones
print("Escribiendo datos procesados...")
sales_processed.write.format("parquet").mode("append").partitionBy("year", "month").save(
    sales_processed_path
)

print(f"Datos escritos en: {sales_processed_path}")

# Actualizar estado de procesamiento
processing_record: dict = {
    "timestamp": datetime.now().isoformat(),
    "source": parquet_file,
    "destination": sales_processed_path,
    "total_records_processed": sales_processed.count(),
    "partitions": ["year", "month"],
    "status": "success",
}

current_status["last_processed_month"] = "2018-03"  # Último mes en los datos de muestra
current_status["history"] = current_status.get("history", [])
current_status["history"].append(processing_record)

# Guardar estado
write_processing_status(current_status)
```

### Paso 3: Procesar Datos Dimensionales (Empleados y Clientes)

Ahora procesamos las tablas dimensionales. Aunque son más pequeñas y estáticas, las enriquecemos con información derivada.

```python
# Agregar columna de antigüedad (días desde contratación hasta hoy)
from pyspark.sql.functions import current_date, datediff

# Procesar Empleados
employees_processed: DataFrame = dfEmpleados.withColumn(
    "DaysEmployed", datediff(current_date(), col("HireDate"))
).withColumn("YearsEmployed", col("DaysEmployed") / 365.0)

# En este caso hacemos coalesce(1) porque los datos de empleados suelen ser pequeños
# y queremos un solo fichero de salida
employees_processed_path: str = f"{PROCESSED_PATH}/employees_processed"
employees_processed.coalesce(1).write.format("parquet").mode("overwrite").save(
    employees_processed_path
)
print(f"Empleados procesados: {employees_processed.count()} registros")

# Procesar Clientes
customers_clean: DataFrame = dfClientes.dropDuplicates(["CustomerID"])

# Agregar información derivada: número de compras (desde sales)
customer_purchase_count: DataFrame = (
    sales.groupBy("CustomerID").count().withColumnRenamed("count", "TotalPurchases")
)

customers_processed: DataFrame = customers_clean.join(
    customer_purchase_count, on="CustomerID", how="left"
).fillna(0, subset=["TotalPurchases"])

customers_processed_path: str = f"{PROCESSED_PATH}/customers_processed"
customers_processed.coalesce(1).write.format("parquet").mode("overwrite").save(
    customers_processed_path
)
print(f"Clientes procesados: {customers_processed.count()} registros")
```

### Paso 4: Validación de Calidad de Datos

Ejecutamos checks de validación para asegurar que los datos procesados cumplen estándares mínimos de calidad.

```python
from typing import Any

from pyspark.sql.functions import count, isnull, when


def validate_sales_data(df: DataFrame) -> dict[str, Any]:
    """Ejecuta validaciones sobre el DataFrame de ventas procesadas."""
    validations: dict[str, Any] = {}

    # 1. Verificar que no hay SalesID duplicados
    unique_sales: int = df.select("SalesID").distinct().count()
    total_sales: int = df.count()
    validations["duplicates_check"] = {
        "expected": total_sales,
        "actual": unique_sales,
        "status": "PASS" if unique_sales == total_sales else "FAIL",
    }

    # 2. Verificar que TotalPrice > 0
    valid_prices: int = df.filter(col("TotalPrice") > 0).count()
    validations["positive_price_check"] = {
        "total_records": total_sales,
        "valid_prices": valid_prices,
        "status": "PASS" if valid_prices == total_sales else "FAIL",
    }

    # 3. Verificar nulls críticos
    null_checks: Row = df.select(
        count(when(isnull(col("SalesID")), 1)).alias("null_SalesID"),
        count(when(isnull(col("SalesDate")), 1)).alias("null_SalesDate"),
        count(when(isnull(col("TotalPrice")), 1)).alias("null_TotalPrice"),
    ).collect()[0]

    validations["null_check"] = dict(null_checks.asDict())
    validations["null_check"]["status"] = (
        "PASS" if all(v == 0 for k, v in null_checks.asDict().items()) else "FAIL"
    )

    return validations


# Ejecutar validaciones
sales_validation: dict[str, Any] = validate_sales_data(sales_processed)
print("=== Validación de Datos Procesados ===")
for check_name, check_result in sales_validation.items():
    print(f"\n{check_name}:")
    print(f"  {check_result}")
```

### Paso 5: Lectura Incremental (Continuidad del Procesamiento)

En ejecuciones posteriores, podemos detectar el último mes procesado y solo procesar nuevos datos. Esto optimiza el pipeline incrementalmente.

```python
from pyspark.sql.dataframe import DataFrame


def get_next_period_to_process(last_processed: str | None) -> str:
    """
    Determina el siguiente período a procesar basado en el último procesado.
    Ej: Si last_processed="2018-03", devuelve "2018-04"
    """
    if last_processed is None:
        return "2018-01"  # Comenzar desde el inicio

    year, month = last_processed.split("-")
    year, month = int(year), int(month)

    # Pasar al siguiente mes
    month += 1
    if month > 12:
        month = 1
        year += 1

    return f"{year}-{month:02d}"


def read_partition_for_period(base_path: str, year: int, month: int) -> DataFrame | None:
    """Lee datos desde un directorio particionado por year/month."""
    partition_path: str = f"{base_path}/year={year}/month={month}"
    try:
        df: DataFrame = spark.read.format("parquet").load(partition_path)
        return df
    except Exception as e:
        print(f"No hay datos para {year}-{month:02d}: {e}")
        return None


# Demostración: calcular qué período sería el siguiente
next_period: str = get_next_period_to_process(current_status.get("last_processed_month"))
print(f"Siguiente período a procesar: {next_period}")

# Leer datos procesados para inspección
print("\n=== Inspección de Datos Procesados ===")
try:
    sales_processed_read: DataFrame = spark.read.format("parquet").load(sales_processed_path)
    print(f"Total de registros en processed/sales_processed: {sales_processed_read.count()}")

    # Mostrar particiones disponibles
    print("\nMeses disponibles en processed:")
    months_data: DataFrame = (
        sales_processed_read.select("year", "month").distinct().orderBy("year", "month")
    )
    months_data.show(10)
except Exception as e:
    print(f"Error al leer datos procesados: {e}")
```

## Agregaciones Silver: Análisis por Empleado, Cliente y Producto

Creamos tablas de agregación que facilitan análisis y reportes sin necesidad de hacer joins complejos cada vez.


```python
from pyspark.sql.functions import avg as sql_avg
from pyspark.sql.functions import count as sql_count
from pyspark.sql.functions import max as sql_max
from pyspark.sql.functions import min as sql_min
from pyspark.sql.functions import sum as sql_sum

# Leer datos procesados para las agregaciones
sales_for_agg: DataFrame = spark.read.format("parquet").load(sales_processed_path)
sales_for_agg.cache()  # Cachear para múltiples usos. Esperemos que no sea muy grande.

print("=" * 80)
print("AGREGACIONES SILVER - ANÁLISIS POR EMPLEADO, CLIENTE Y PRODUCTO")
print("=" * 80)
```

### 1. Total de Ventas por Empleado (Global)


```python
sales_by_employee: DataFrame = (
    sales_for_agg.groupBy("SalesPersonID", "EmployeeFirstName", "EmployeeLastName")
    .agg(
        sql_count("SalesID").alias("TotalTransactions"),
        sql_sum("TotalPrice").alias("TotalSalesAmount"),
        sql_sum("Quantity").alias("TotalQuantity"),
        sql_avg("TotalPrice").alias("AvgSalesAmount"),
        sql_sum("Commission").alias("TotalCommissions"),
        sql_avg("Discount").alias("AvgDiscount"),
    )
    .orderBy(col("TotalSalesAmount").desc())
)

sales_by_employee_path: str = f"{PROCESSED_PATH}/sales_by_employee"
sales_by_employee.coalesce(1).write.format("parquet").mode("overwrite").save(sales_by_employee_path)

print("\n1. VENTAS POR EMPLEADO (Global)")
print(f"   Guardado en: {sales_by_employee_path}")
sales_by_employee.show(10, truncate=False)
```

### 2. Ventas por Empleado y Período (Año-Mes)


```python
sales_by_employee_period: DataFrame = (
    sales_for_agg.groupBy("year", "month", "SalesPersonID", "EmployeeFirstName", "EmployeeLastName")
    .agg(
        sql_count("SalesID").alias("TransactionsInPeriod"),
        sql_sum("TotalPrice").alias("SalesAmountInPeriod"),
        sql_sum("Quantity").alias("QuantityInPeriod"),
        sql_sum("Commission").alias("CommissionInPeriod"),
    )
    .orderBy(col("year").desc(), col("month").desc(), col("SalesAmountInPeriod").desc())
)

sales_by_employee_period_path: str = f"{PROCESSED_PATH}/sales_by_employee_period"
sales_by_employee_period.write.format("parquet").mode("overwrite").partitionBy(
    "year", "month"
).save(sales_by_employee_period_path)

print("\n2. VENTAS POR EMPLEADO Y PERÍODO (Año-Mes)")
print(f"   Guardado en: {sales_by_employee_period_path}")
sales_by_employee_period.filter(col("year") == 2018).filter(col("month") == 3).show(
    10, truncate=False
)
```

### 3. Total de Compras por Cliente (Global)


```python
sales_by_customer: DataFrame = (
    sales_for_agg.groupBy("CustomerID")
    .agg(
        sql_count("SalesID").alias("TotalPurchases"),
        sql_sum("TotalPrice").alias("TotalSpent"),
        sql_sum("Quantity").alias("TotalItemsBought"),
        sql_avg("TotalPrice").alias("AvgPurchaseAmount"),
        sql_max("SalesDate").alias("LastPurchaseDate"),
        sql_min("SalesDate").alias("FirstPurchaseDate"),
    )
    .orderBy(col("TotalSpent").desc())
)

sales_by_customer_path: str = f"{PROCESSED_PATH}/sales_by_customer"
sales_by_customer.coalesce(1).write.format("parquet").mode("overwrite").save(sales_by_customer_path)

print("\n3. COMPRAS POR CLIENTE (Global)")
print(f"   Guardado en: {sales_by_customer_path}")
sales_by_customer.show(10, truncate=False)
```

### 4. Compras por Cliente y Período (Año-Mes)


```python
sales_by_customer_period: DataFrame = (
    sales_for_agg.groupBy("year", "month", "CustomerID")
    .agg(
        sql_count("SalesID").alias("PurchasesInPeriod"),
        sql_sum("TotalPrice").alias("SpentInPeriod"),
        sql_sum("Quantity").alias("ItemsInPeriod"),
        sql_avg("TotalPrice").alias("AvgPurchaseInPeriod"),
    )
    .orderBy(col("year").desc(), col("month").desc(), col("SpentInPeriod").desc())
)

sales_by_customer_period_path: str = f"{PROCESSED_PATH}/sales_by_customer_period"
sales_by_customer_period.write.format("parquet").mode("overwrite").partitionBy(
    "year", "month"
).save(sales_by_customer_period_path)

print("\n4. COMPRAS POR CLIENTE Y PERÍODO (Año-Mes)")
print(f"   Guardado en: {sales_by_customer_period_path}")
sales_by_customer_period.filter(col("year") == 2018).filter(col("month") == 3).show(
    10, truncate=False
)
```

### 5. Productos Más Vendidos por Período (Año-Mes)


```python
products_by_period: DataFrame = (
    sales_for_agg.groupBy("year", "month", "ProductID")
    .agg(
        sql_count("SalesID").alias("TimesOld"),
        sql_sum("Quantity").alias("QuantitySold"),
        sql_sum("TotalPrice").alias("TotalRevenue"),
        sql_avg("Discount").alias("AvgDiscount"),
    )
    .orderBy(col("year").desc(), col("month").desc(), col("TotalRevenue").desc())
)

products_by_period_path: str = f"{PROCESSED_PATH}/products_by_period"
products_by_period.write.format("parquet").mode("overwrite").partitionBy("year", "month").save(
    products_by_period_path
)

print("\n5. PRODUCTOS MÁS VENDIDOS POR PERÍODO (Año-Mes)")
print(f"   Guardado en: {products_by_period_path}")
products_by_period.filter(col("year") == 2018).filter(col("month") == 3).show(10, truncate=False)
```

### 6. Productos Más Vendidos Globales


```python
products_global: DataFrame = (
    sales_for_agg.groupBy("ProductID")
    .agg(
        sql_count("SalesID").alias("TimesSold"),
        sql_sum("Quantity").alias("TotalQuantitySold"),
        sql_sum("TotalPrice").alias("TotalRevenue"),
        sql_avg("Discount").alias("AvgDiscount"),
        sql_max("SalesDate").alias("LastSaleDate"),
    )
    .orderBy(col("TotalRevenue").desc())
)

products_global_path: str = f"{PROCESSED_PATH}/products_global"
products_global.coalesce(1).write.format("parquet").mode("overwrite").save(products_global_path)

print("\n6. PRODUCTOS MÁS VENDIDOS (Global)")
print(f"   Guardado en: {products_global_path}")
products_global.show(10, truncate=False)
```

### 7. Productos por Cliente (Preferencias de Compra)


```python
products_by_customer: DataFrame = (
    sales_for_agg.groupBy("CustomerID", "ProductID")
    .agg(
        sql_count("SalesID").alias("PurchaseCount"),
        sql_sum("Quantity").alias("TotalQuantity"),
        sql_sum("TotalPrice").alias("TotalSpent"),
        sql_max("SalesDate").alias("LastPurchaseDate"),
    )
    .orderBy(col("CustomerID"), col("TotalSpent").desc())
)

products_by_customer_path: str = f"{PROCESSED_PATH}/products_by_customer"
products_by_customer.write.format("parquet").mode("overwrite").save(products_by_customer_path)

print("\n7. PRODUCTOS POR CLIENTE (Preferencias de Compra)")
print(f"   Guardado en: {products_by_customer_path}")
products_by_customer.limit(15).show(15, truncate=False)
```

### 8. Productos por Vendedor (Especialización)


```python
products_by_salesperson: DataFrame = (
    sales_for_agg.groupBy("SalesPersonID", "EmployeeFirstName", "EmployeeLastName", "ProductID")
    .agg(
        sql_count("SalesID").alias("SalesCount"),
        sql_sum("Quantity").alias("TotalQuantity"),
        sql_sum("TotalPrice").alias("TotalRevenue"),
        sql_avg("Discount").alias("AvgDiscount"),
    )
    .orderBy(col("SalesPersonID"), col("TotalRevenue").desc())
)

products_by_salesperson_path: str = f"{PROCESSED_PATH}/products_by_salesperson"
products_by_salesperson.write.format("parquet").mode("overwrite").save(products_by_salesperson_path)

print("\n8. PRODUCTOS POR VENDEDOR (Especialización)")
print(f"   Guardado en: {products_by_salesperson_path}")
products_by_salesperson.limit(15).show(15, truncate=False)
```

### 9. Clientes por Vendedor (Carteras de clientes)


```python
customers_by_salesperson: DataFrame = (
    sales_for_agg.groupBy("SalesPersonID", "EmployeeFirstName", "EmployeeLastName", "CustomerID")
    .agg(
        sql_count("SalesID").alias("TransactionsWithCustomer"),
        sql_sum("TotalPrice").alias("TotalRevenueFromCustomer"),
        sql_max("SalesDate").alias("LastSaleToCustomer"),
    )
    .groupBy("SalesPersonID", "EmployeeFirstName", "EmployeeLastName")
    .agg(
        sql_count("CustomerID").alias("TotalUniqueCustomers"),
        sql_sum("TotalRevenueFromCustomer").alias("TotalRevenueFromAllCustomers"),
        sql_avg("TotalRevenueFromCustomer").alias("AvgRevenuePerCustomer"),
        sql_max("LastSaleToCustomer").alias("LastSaleDate"),
    )
    .orderBy(col("TotalRevenueFromAllCustomers").desc())
)

customers_by_salesperson_path: str = f"{PROCESSED_PATH}/customers_by_salesperson"
customers_by_salesperson.coalesce(1).write.format("parquet").mode("overwrite").save(
    customers_by_salesperson_path
)

print("\n9. CLIENTES POR VENDEDOR (Carteras)")
print(f"   Guardado en: {customers_by_salesperson_path}")
customers_by_salesperson.show(10, truncate=False)
```

### 10. Análisis de Desempeño: Vendedor-Período-Cliente


```python
performance_detailed: DataFrame = (
    sales_for_agg.groupBy(
        "year",
        "month",
        "SalesPersonID",
        "EmployeeFirstName",
        "EmployeeLastName",
        "CustomerID",
    )
    .agg(
        sql_count("SalesID").alias("Transactions"),
        sql_sum("Quantity").alias("Quantity"),
        sql_sum("TotalPrice").alias("Revenue"),
        sql_sum("Commission").alias("Commission"),
        sql_avg("Discount").alias("AvgDiscount"),
    )
    .orderBy(
        col("year").desc(),
        col("month").desc(),
        col("SalesPersonID"),
        col("Revenue").desc(),
    )
)

performance_detailed_path: str = f"{PROCESSED_PATH}/performance_detailed"
performance_detailed.write.format("parquet").mode("overwrite").partitionBy("year", "month").save(
    performance_detailed_path
)

print("\n10. DESEMPEÑO DETALLADO (Vendedor-Período-Cliente)")
print(f"   Guardado en: {performance_detailed_path}")
performance_detailed.filter(col("year") == 2018).filter(col("month") == 3).limit(15).show(
    15, truncate=False
)
```

### Resumen del Proceso ETL Landing → Processed

**Características clave implementadas:**

1. **Limpieza de datos**: Eliminación de duplicados, validación de tipos
2. **Enriquecimiento**: Joins con tablas dimensionales, cálculos derivados (comisiones)
3. **Particionamiento**: Organización por Year/Month para optimizar lecturas
4. **Tracking de estado**: Archivo `_processing_status.json` para mantener historial
5. **Validación**: Checks de calidad (duplicados, nulls, valores válidos)
6. **Incrementalidad**: Capacidad de procesar solo nuevos períodos sin reprocesar

**Próximas ejecuciones:**
- El pipeline verificaría `current_status["last_processed_month"]`
- Calcularía el siguiente período con `get_next_period_to_process()`
- Solo procesaría datos nuevos
- Actualizaría el estado al finalizar

**Escalabilidad para producción:**
- Usar Airflow/scheduling para ejecuciones automáticas
- Configurar alertas si validaciones fallan
- Implementar rollback de particiones en caso de error
- Monitorear tiempos de ejecución y recursos
